## Khám phá dữ liệu

Ở phần này, chúng ta sẽ khám phá dữ liệu ở các tiêu chí:
- Phân phối giá trị của các biến đơn.
- Mối liên hệ giữa các biến.


```python
df_book.head()
```

### Top author

```python
plt.figure(figsize=(10,10))
plot = sns.countplot(y = "Author_Name", data = df_book, order = df_book[df_book.Author_Name != "Anonymous"]['Author_Name'].value_counts().iloc[:10].index, palette = "Set3")
```

Top 10 tác giả có nhiều tác phẩm nhất, mỗi tác giả đều có trên 300 đầu sách.

### Top rating counts

```python
most_rated = df_book.sort_values('Rating_Counts', ascending = False).head(10).set_index('Book_Title')
plt.figure(figsize=(15,10))
sns.barplot(most_rated['Rating_Counts'], most_rated.index, palette='Set3')
```

Điều thú vị ở top các cuốn sách có số lượng rating nhiều nhất đều là các phần của Harry Potter.

### Language

```python
plt.figure(1, figsize=(25,10))
plt.title("Language")
sns.countplot(x = "Language", order=df_book['Language'].value_counts().index[0:10] ,data=df_book)
```

Hầu hết các cuốn sách trên thư viện goodreads đều là các đầu sách tiếng Anh.

### Rating Distribution

```python
plt.figure(figsize=(10,10))
plot = sns.countplot(y = "Rating", data = df_book, order = df_book['Rating'].value_counts().iloc[:10].index, palette = "Set3")
```

```python
sns.distplot(df_book['Rating'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
```

Phần lớn các cuốn sách có phân phối điểm rating từ 3 đến 5 có hình dạng của phân phối chuẩn.

### Num pages distribution

```python
sns.distplot(df_book['Num_Pages'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
```

Tập dữ liệu có một vài cuốn sách có số trang lớn hơn 10000, đây hầu hết là các cuốn từ điển hoặc bách khoa toàn thư. Chúng ta xóa đi các cuốn sách có số trang lớn hơn 10000 để có cái nhìn trực quan.

```python
df_book = df_book.drop(df_book[df_book.Num_Pages > 10000].index.to_list())
```

```python
ax = sns.distplot(df_book['Num_Pages'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
ax.set_title("Num pages < 10000")
```

```python
ax = sns.distplot(df_book[(df_book.Num_Pages < 2000) & (df_book.Num_Pages > 0)]['Num_Pages'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
ax.set_title("Num pages < 2000")
```

Tập dữ liệu có số trang của các cuốn sách phân phối từ 0 đến 1000 trang.

Biểu đồ bên dưới là top các cuốn sách có số trang cao nhất.

```python
top_pages = df_book.sort_values(by='Num_Pages', ascending=False).head(10)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(5, 20))

color = sns.color_palette("Set2")
ax = sns.barplot(x="Num_Pages", y="Book_Title", data=top_pages, palette=color)

for i in ax.patches:
    ax.text(i.get_width() + .05, i.get_y() + 0.5, str(i.get_width()), fontsize = 10, color = 'k')
plt.show()
```

### Rating vs Num Pages

```python
ax = sns.relplot(data=df_book, x="Rating", y="Num_Pages", color = '#95a3c3', sizes=(100, 200), height=7, marker='o')
```

```python
req = df_book.groupby(pd.cut(df_book['Rating'], [0,1,2,3,4,5]))
req = req[['Num_Pages']]
req.sum().reset_index()
```

Các cuốn sách có số trang cao hơn 2000 có mức rating từ 3 tới 5 (trừ 0 là các cuốn sách chưa có rating từ độc giả).

Tập dữ liệu bị imbalanced, với các số lượng các đầu sách tỉ lệ thuận với điểm Rating. Chúng ta cần xử lý phần này ở phần mô hình.

### Rating Counts vs Num Pages

```python
ax = sns.relplot(data=df_book, x="Num_Pages", y="Rating_Counts", color = '#95a3c3', sizes=(100, 200), height=7, marker='o')
```

```python
req = df_book.groupby(pd.cut(df_book['Num_Pages'], [0,1000,2000,3000,4000,5000]))
req = req[['Rating_Counts']]
req.sum().reset_index()
```

Các đầu sách có số trang càng ít thì số Rating_Counts càng cao. Chứng tỏ là các đầu sách có nột dung ngắn có xu hướng có nhiều người đọc hơn là các cuốn sách có nội dung dài.

### Rating Counts vs Rating

```python
ax = sns.relplot(data=df_book, x="Rating", y="Rating_Counts", color = '#95a3c3', sizes=(100, 200), height=7, marker='o')
```

```python
req = df_book.groupby(pd.cut(df_book['Rating'], [0,1,2,3,4,5]))
req = req[['Rating_Counts']]
req.sum().reset_index()
```

Điểm Rating và Rating_Counts tỉ lệ thuận với nhau. Điều này đúng với thực tế, các đầu sách hay có nhiều người đọc hơn và ngược lại.
